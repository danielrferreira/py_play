# Linear Transformations to change images

```python
import numpy as np
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
image_raw = Image.open("/Users/danielferreira/Documents/Python_training/cube/cubo.jpeg")
image_array = np.array(image_raw)
print(image_array.shape)
print(image_array.shape[0]*image_array.shape[1])
plt.imshow(image_array)
```

```python
new_image = np.empty_like(image_array)
```

This matrix can be used to strech and shrink:

```python
T = np.array([[0.75,0],[0,0.75]])
```

If you to rotate the image, you can use rotation matrix:

```python
teta = 30
teta_rads = math.radians(teta)
T = np.array([[math.cos(teta_rads),-math.sin(teta_rads)],[math.sin(teta_rads),math.cos(teta_rads)]])
```

```python
len_x = image_array.shape[0]
len_y = image_array.shape[1]
for x in range(len_x-1):
    for y in range(len_y-1):
        v = np.array([[x],[y]])
        new_v = np.matmul(T,v)
        if (((int(new_v[0,0])>=0) & (int(new_v[0,0])<len_x-1) ) & ( (int(new_v[1,0])>=0) & (int(new_v[1,0])<len_y-1))):
            new_image[int(new_v[0,0]),int(new_v[1,0])] = image_array[x,y]

plt.close()
plt.imshow(new_image)
```
